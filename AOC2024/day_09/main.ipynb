```python
class DiskSort():
    def __init__ (self,txt):
        print(len(txt))
        self.disk = []
        
        for i in range(0, len(txt)):
            if (i%2==0): # file
                # print([int((i)/2) for x in range(int(txt[i]))]
                self.disk += [str(int((i)/2)) for x in range(int(txt[i]))]
            else: # space
                # print(['.' for x in range(int(txt[i]))])
                self.disk += ['.' for x in range(int(txt[i]))]
        
        pointer = len(self.disk)-1
        disk = list(self.disk)
        for i in range(0, len(disk)):
            if disk[i] == '.':
                while(disk[pointer]=='.'):
                    pointer-=1
                disk[i] = disk[pointer]
                disk[pointer] = '.'
                pointer-=1
            if i == pointer: break
        self.disk = disk
        # self.compacted_disk = ''.join(disk)
        
def calculate_checksum(disk):
    return sum([int(disk[i]) * i for i in range(len(disk)) if disk[i] != '.'])
    # str = txt
    # return sum([int(str[i])*i for i in disk])


res = DiskSort(input.strip())
print(res.disk)
# print(res.compacted_disk)
print(calculate_checksum(res.disk))
# assert res.compacted_disk == "0099811188827773336446555566.............."

# 90167081070 -> too low?? (I failed to consider single digit indexes)
# 6359213660505

print
```
