```python
# Reference: HyperNeutrino https://www.youtube.com/watch?v=kHIWvxRWQ9k

class LANParty:
    def __init__(self,txt):
        self.g = {}
        self.loops = set()
        for line in txt.splitlines():
            # print(line)
            u,v =  line.split("-")
            # for u,v in line.split("-"):
            # print(u,v)
            if u not in self.g:
                self.g[u] = []
            self.g[u].append(v)

            if v not in self.g:
                self.g[v] = []
            self.g[v].append(u)

    # FIND 3-SIZED LOOP
    def find_loops(self):
        loops = set()
        for u in self.g:
            for v in self.g[u]:
                for w in self.g[v]:
                    if u in self.g[w]:
                        loops.add(tuple(sorted([u,v,w])))
        return loops
    
    def find_t(self,groups):
        c = 0
        for group in groups:
            for name in group:
                if name.startswith("t"):
                    c+=1
                    break
        return c
    
    # THIS IS TO FIND A N-SIZE LINE! NOT LOOP.
    def find_connections(self,v,n,xs):
        connections = []
        # print(f"v,n,xs {v,n,xs}")
        for u in self.g[v]:
            # print(f"u: {u}")
            if u in xs: continue
            if n == 1 and xs[0] in self.g[u]:
                print(f"xs[0] {xs[0]} | u {u} | self.g[u] {self.g[u]}")
                connections.append(u)
                continue
            # not sure about the n > 1 logic...
            nxs = xs + [u]
            # print(f"nxs{nxs}")
            sub_connections = self.find_connections(u,n-1,nxs)
            if sub_connections != []:
                # print(f"subc: {sub_connections}")
                for sc in sub_connections:                
                    connections.append(nxs + sc)
        return connections

    def expand_loop(self,curr_node,curr_loop):
        # base case - assumes curr_loop is a loop by parameter
        sorted_curr_loop = tuple(sorted(curr_loop))
        if sorted_curr_loop in self.loops: return # if it's already there, ignore
        self.loops.add(sorted_curr_loop)
        for neighbor in self.g[curr_node]:
            # if neighbor is already in the set, continue
            if neighbor in curr_loop: continue
            # if not all nodes within the set are connected to the neighbor, continue
            if not all([neighbor in self.g[u] for u in curr_loop]): continue
            # if it is connected to all node, means curr_loop + neighbor is a new loop
            # we can search for a bigger loop with it()
            self.expand_loop(neighbor,{*curr_loop, neighbor})

res = LANParty(input)
# print(res.g)
# print()
groups = res.find_loops()
print("part1:", res.find_t(groups))
for u in res.g:
    res.expand_loop(u,{u}) # begin with a single node

print("part2:",','.join(max(res.loops, key=len)))
```
