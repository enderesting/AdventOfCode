```python
with open('input.txt', 'r') as file:
    input = file.read()

import re
```

```python
test_input = """\
3   4
4   3
2   5
1   3
3   9
3   3
"""

pattern = r"(?P<first>\d*)   (?P<last>\d*)"

def find_pairs(input):
    regex = re.compile(pattern, re.MULTILINE)
    matches = regex.finditer(input)
    l = []
    for match in matches:
        l.append([int(match["first"]), int(match["last"])])
    print(l)

    res = [[i for i,j in l], [j for i,j in l]]
    print(res)
    # zip(res)
    # res = list(zip(*l))
    res[0].sort()
    res[1].sort()
    print(res)
    
    ans = []
    for i in range(0,len(res[0])):
        ans.append(abs(res[0][i]-res[1][i]))
    print(sum(ans))




find_pairs(test_input)

find_pairs(input)

```

```python
def find_similarity(input):
    regex = re.compile(pattern, re.MULTILINE)
    matches = regex.finditer(input)
    l = []
    for match in matches:
        l.append([int(match["first"]), int(match["last"])])
    print(l)

    res = [[i for i,j in l], [j for i,j in l]]
    print(res)

    ans = []
    sum = 0
    for i in range(0,len(res[0])):
        sum += (res[0][i]*res[1].count(res[0][i]))
    print(sum)

find_similarity(test_input)
find_similarity(input)
```

```python
test_input = """\
1abc2
pqr3stu8vwx
a1b2c3d4e5f
treb7uchet
"""

pattern = r"^\D*(?P<first>\d).*?(?P<last>\d)?\D*$"
# ^         - match start
# \D*       - not digit, * >= 0 occurrances
# (?P<first>\d)     - (?P<>...) trying to find matching ... which is a group that can be called by first
# .*?               - . matches any character, *? lazy quantifier, matches as few as possible
# (?P<last>\d)?     - match a digit, ? at the end makes it optional
# $         - match end

regex = re.compile(pattern, re.MULTILINE)
print(regex)

def find_number():
    matches = regex.finditer(test_input)
    list = []
    for match in matches:
        list.append(int( match["first"] + (match["last"] or match["first"])))
    return list

m = find_number()

assert m == [12, 38, 15, 77]
assert sum(m) == 142
```
