```python
with open('input.txt', 'r') as file:
    input = file.read()

import re

test_input = """\
xmul(2,4)&mul[3,7]!^don't()_mul(5,5)+mul(32,64](mul(11,8)undo()?mul(8,5))
"""
```

```python
pattern_first = r""
pattern = r".*?mul\((?P<x>\d{1,3}),(?P<y>\d{1,3})\).*?"

def multiply_jumbled(txt):
    regex = re.compile(pattern, re.MULTILINE)
    matches = regex.finditer(txt)
    print(sum([int(match["x"]) * int(match["y"]) for match in matches]))

multiply_jumbled(test_input)
multiply_jumbled(input)
```

```python
pattern_first = r"^.*?mul\((?P<x>\d{1,3}),(?P<y>\d{1,3})\)"
pattern_rest = r"(?=(do\(\)).*?mul\((?P<x>\d{1,3}),(?P<y>\d{1,3})\))"
def multiply_with_do(txt):
    regex = re.compile(pattern_first,re.MULTILINE)
    matches = regex.finditer(txt)
    first_sums = sum([int(match["x"]) * int(match["y"]) for match in matches])
    regex2 = re.compile(pattern_rest,re.MULTILINE)
    matches = regex2.finditer(txt)
    # print([match["y"] for match in matches])
    rest_sums = sum([int(match["x"]) * int(match["y"]) for match in matches])
    print(first_sums+rest_sums)

multiply_with_do(test_input)
multiply_with_do(input)
```

```python
# ref: nitekat1124
# https://github.com/nitekat1124/advent-of-code-2024/blob/main/solutions/day03.py
pattern = r"(mul\((\d{1,3}),(\d{1,3})\)|do\(\)|don't\(\))"

def multiply_with_instructions(txt):
    instructions = re.findall(pattern,txt)
    enabled = True
    sum = 0
    print(instructions)

    for inst in instructions:
        match inst[0]:
            case "do()":
                enabled = True
            case "don't()":
                enabled = False
            case _ if enabled:
                sum += int(inst[1]) *int(inst[2])
    print(sum)

multiply_with_instructions(test_input)
multiply_with_instructions(input)
```
