```python
class TowelDisplay:
    def __init__(self,txt):
        a,b = txt.split("\n\n")
        self.towels = a.split(", ")
        self.designs = b.splitlines()
        print(self.towels)
        print(self.designs)
    
    @cache
    def match_design(self,design):
        # print("current design:", design)
        found = 0
        for i in range(len(design),0,-1):
            sub_design = design[:i]
            # print("     sub design",sub_design)
            if sub_design in self.towels:
                remaining = design[i:]
                # print(f"     {sub_design}, next section:{remaining}")
                if remaining == "":
                    found += 1
                found += self.match_design(remaining)
                # if found: break # part 1
        return found

res = TowelDisplay(input)
# print(sum([res.match_design(d) for d in res.designs])) # 1.4s
p1,p2 = 0,0
for d in res.designs:
    found_count = res.match_design(d)
    p2 += found_count
    if found_count != 0:
        p1+=1
print(f"part1: {p1}")
print(f"part2: {p2}")
```
